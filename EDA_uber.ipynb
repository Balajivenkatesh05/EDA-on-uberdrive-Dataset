```python
#Q15
data['START*'].value_counts().head(1)
```

```python
#alternative way
starting_point = data['START*'].dropna()
sdf =pd.DataFrame(starting_point.value_counts())
sdf
sdf = sdf.reset_index()
sdf = sdf.rename(columns={'index':'Starting Point', 'START*':'Count'})
sdf
```

```python
sdf.loc[sdf['Count'] == max(sdf['Count'])]
```

```python
#Q16
stop_point = data['STOP*'].dropna()
stdf =pd.DataFrame(stop_point.value_counts())
stdf
stdf = stdf.reset_index()
stdf = stdf.rename(columns={'index':'Stopping Point', 'STOP*':'Count'})
stdf
```

```python
stdf.loc[stdf['Count'] == max(stdf['Count'])]
```

```python
#Q17
data2 = data.dropna()
mfdf = pd.DataFrame(data2.groupby(['START*','STOP*']).size())
mfdf = mfdf.rename(columns = {0: 'Count'})
mfdf
```

```python
mfdf.loc[mfdf['Count'] == max(mfdf['Count'])]
```
